```python
df = pd.read_csv('cleaned_books.csv')
df = df.drop('Unnamed: 0', 1)
df.head()
```

```python
df.info()
```

```python
# check the null values
df.isnull().sum()
```

```python
# Make a list of ISBN-10 for API later
isbn_list = df['primary_isbn10'].to_list()
isbn_list
```

```python
# API fetching
url = 'https://www.goodreads.com/book/review_counts.json?key=&isbns='
bad_isbns = []
dict_api = []
for countr, i in enumerate(isbn_list):
    if countr % 10 == 0:
        print(countr)
    try: 
        response = requests.get(url + str(i))
        dict_api.append(response.json())
    except:
        bad_isbns.append(i)
```

```python
# review the result from API fetching
dict_api
```

```python
# check to see how many data points left after eliminating the bad ISBNs
len(dict_api)
```

```python
# convert to a dataframe 
review_stats = pd.concat([pd.DataFrame(i['books']) for i in dict_api])
```

```python
# review the dataframe
review_stats.head(10)
```

```python
# review the general info of the dataframe
review_stats.info()
```

```python
# Change the data type for the 'average_rating' column
review_stats['average_rating'] = review_stats['average_rating'].astype('float')
```

```python
review_stats.info()
```

```python
# Drop unnecessary columns
review_stats = review_stats.drop(columns=['id', 'isbn13'], axis=1).copy()
```

```python
# Check for null values
review_stats.isnull().sum()
```

```python
# Check for duplicated values based on ISBNs
review_stats['isbn'].duplicated().sum()
```

```python
# Drop duplicated values
review_stats.drop_duplicates(subset ='isbn', inplace = True) 
```

```python
review_stats.info()
```

```python
# Save the dataframe to a CSV file
review_stats.to_csv('review_stats.csv')
```
